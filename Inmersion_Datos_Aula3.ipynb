Se almacena en una variable y se agregan estas dos columnas al DataFrame

```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles['Moneda']= valor[0] #Columna 0 -> Moneda en DF
inmuebles['Precio']=valor[1] #Columna 1 -> Precio en DF
inmuebles.sample(3)
```

Convertir los datos de la columna Precio de str a numerico


*   Quitar el separador "." utilizando .replace (regex=True se agrega por actualizacion de python)
*   Convertir str a float, utilizando **astype** en un nuevo atributo Precio Millon



```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','',regex=True)
```

```python
inmuebles['Precio_Millon'] = inmuebles.Precio.astype(float)/1000000
inmuebles.info()
```

La manera para solucionar el segundo Desafio del Aula 2:

Se suman todos los inmuebles que pertenecen a cada Barrio, tambien su precio y se van a divir entre la sumatoria de toda la superficies que pertenecen a su Barrio, y asi para calcular el valor especifico del m2 de cada uno de los Barrios

```python
datos_barrio = inmuebles.groupby('Barrio').sum()
datos_barrio
```

**Nuevo DataSet**

Apartir de estas sumas se agregara una nueva columna a este data set ***datos_barrio*** en donde se volvera a calcular el Precio/m2

```python
datos_barrio['Valor_m2_Barrio'] = datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

Se mapearan estos datos

Primero se almacena en un diccionario, teniendo como Key: Barrio, Valor: Precio x m2 (del dataset anterior)

```python
m2_barrio = dict(datos_barrio['Valor_m2_Barrio'])
```

Ahora se busca llevar este nuevo valor al DataSet orginala.

Se sustituye *Barrio*  por *Valor_m2_Barrio*, duplicando una columna (Barrios) y se asigna a la del m2

```python
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['Valor_m2_Barrio'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio) #map toma los valor de dict
inmuebles.head(5)
```

Ahora se busca visualizar el Top 10 de Barrios que contienen más inmuebles a la venta

```python
top_barrios = inmuebles['Barrio'].value_counts()[:10].index #obtiene el top 10 de Barrio
```

Como en el nuevo DataSet ***datos_barrio*** el Barrio aparece como indice, para quitarlo se hace lo siguente

```python
datos_barrio.reset_index(inplace=True)
datos_barrio
```

Se realiza consultas al Nuevo DataSet utilizando ***.query*** para obtener el Top 10 de Barrios

```python
datos_barrio.query('Barrio in @top_barrios')
```
